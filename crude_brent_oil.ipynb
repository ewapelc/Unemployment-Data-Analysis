```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore", "use_inf_as_na")

%matplotlib inline
```

## Data Preprocessing
Reading, processing and summarizing data

```python
import yfinance as yf

# Pobranie danych historycznych z ostatnich 10 lat
crude_oil = yf.download("CL=F", start='2011-01-01', end='2024-03-25')
brent_oil = yf.download("BZ=F", start='2011-01-01', end='2024-03-25')
print(crude_oil)
```

```python
crude_oil.reset_index(inplace=True)
crude_oil.info()
```

```python
brent_oil.reset_index(inplace=True)
brent_oil.info()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(brent_oil.Date, brent_oil.Close, markersize=1, color='blue', label='Brent Oil')
plt.plot(crude_oil.Date, crude_oil.Close, markersize=1, color='red', label='Crude Oil')
plt.title('Kurs ropy na przestrzeni lat')
plt.xlabel('Data')
plt.ylabel('Kurs [USD]')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Reading file and displaying first rows to see the structure of data - unemployment
unemployment = pd.read_excel('files/general_unemployment.xlsx', sheet_name = 1)
unemployment.head()
```

```python
# Viewing columns and their datatypes
unemployment.info()
```

## Exploratory Data Analysis (EDA)

```python
# Cleaning up data - drop null data and irrelevant columns
unemployment = unemployment[unemployment['Atrybut'] != 'n']
unemployment.drop(['Kod', 'Nazwa', 'Jednostka miary', 'Atrybut'], axis = 1, inplace = True)
unemployment.rename(columns={'Wartosc': 'Osoby'}, inplace=True)
unemployment.head()
```

```python
# Preparing month map to create date column
month_mapping = {
    'styczeń' : 1,
    'luty' : 2,
    'marzec' : 3,
    'kwiecień' : 4,
    'maj' : 5,
    'czerwiec' : 6,
    'lipiec' : 7,
    'sierpień' : 8,
    'wrzesień' : 9,
    'październik' : 10,
    'listopad' : 11,
    'grudzień' : 12
}
```

```python
# Applying month map and creating date column
unemployment['Month'] = unemployment['Miesiące'].map(month_mapping)
unemployment.rename(columns = {'Rok' : 'Year'}, inplace = True)
unemployment['Data'] = pd.to_datetime(unemployment[['Year', 'Month']].assign(DAY=1))

# Changing type of number of people to int
unemployment['Osoby'] = unemployment['Osoby'].astype(int)
unemployment.head()
```

```python
# Cleaning up data - drop  irrelevant columns
brent_oil.drop(['Open', 'High', 'Low', 'Adj Close', 'Volume'], axis = 1, inplace = True)
brent_oil.rename(columns={'Close': 'Kurs'}, inplace=True)
brent_oil.head()
```

```python
# Cleaning up data - drop  irrelevant columns
crude_oil.drop(['Open', 'High', 'Low', 'Adj Close', 'Volume'], axis = 1, inplace = True)
crude_oil.rename(columns={'Close': 'Kurs'}, inplace=True)
crude_oil.head()
```

```python
# Calculate mean value of USD/PLN for each month
brent_oil['year'] = brent_oil['Date'].dt.year
brent_oil['month'] = brent_oil['Date'].dt.month
monthly_mean = brent_oil.groupby(['year', 'month'])['Kurs'].mean()

# Merge monthly mean back into the original data
brent_oil = brent_oil.drop(columns=['Date', 'Kurs'])
brent_oil = brent_oil.drop_duplicates(subset=['year', 'month'])
brent_oil = brent_oil.merge(monthly_mean, on=['year', 'month'])
brent_oil['date'] = pd.to_datetime(brent_oil[['year', 'month']].assign(day=1))
brent_oil.head()
```

```python
# Calculate mean value of USD/PLN for each month
crude_oil['year'] = crude_oil['Date'].dt.year
crude_oil['month'] = crude_oil['Date'].dt.month
monthly_mean = crude_oil.groupby(['year', 'month'])['Kurs'].mean()

# Merge monthly mean back into the original data
crude_oil = crude_oil.drop(columns=['Date', 'Kurs'])
crude_oil = crude_oil.drop_duplicates(subset=['year', 'month'])
crude_oil = crude_oil.merge(monthly_mean, on=['year', 'month'])
crude_oil['date'] = pd.to_datetime(crude_oil[['year', 'month']].assign(day=1))
crude_oil.head()
```

```python
fig = plt.figure(figsize = [12,8])

plt.subplot(3, 1, 1)
sns.lineplot(
    x="Data",
    y="Osoby",
    data=unemployment[unemployment['Płeć'] == "ogółem"],
    label="Unemployment"
)

plt.subplot(3, 1, 2)
sns.lineplot(
    x="date",
    y="Kurs",
    data=brent_oil,
    label="Kurs ropy brentowej"
)

plt.subplot(3, 1, 3)
sns.lineplot(
    x="date",
    y="Kurs",
    data=crude_oil,
    label="Kurs ropy naftowej"
)
plt.xlabel("Data")
plt.tight_layout()

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.style.use('seaborn')

# Unemployment data
ax1 = sns.lineplot(
    x="Data",
    y="Osoby",
    data=unemployment[unemployment['Płeć'] == "ogółem"],
    label="Unemployment"
)
ax1.set_xlabel("Data")
ax1.set_ylabel("Osoby", color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Currency data
ax2 = ax1.twinx()
sns.lineplot(
    x="date",
    y="Kurs",
    data=brent_oil,
    ax=ax2,
    color='tab:red',
    label="Kurs ropy brentowej"
)
ax2.set_ylabel("Kurs ropy [USD]", color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Combining legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='lower left')

# Removing duplicate legend
ax2.get_legend().remove()


plt.title("Unemployment and Brent Oil Prices over time")
plt.show()
```
